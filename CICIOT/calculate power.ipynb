```python
import subprocess
import time

def get_gpu_power_usage():
    try:
        smi_output = subprocess.check_output("nvidia-smi --query-gpu=power.draw --format=csv,noheader,nounits", encoding='utf-8', shell=True)
        return float(smi_output.strip())
    except Exception as e:
        print(f"Error getting GPU power usage: {e}")
        return 0.0

def monitor_power_usage(duration=60):
    start_time = time.time()
    power_readings = []

    while time.time() - start_time < duration:
        power_usage = get_gpu_power_usage()
        print(f"Current GPU power usage: {power_usage}W")
        power_readings.append(power_usage)
        time.sleep(1)

    average_power = sum(power_readings) / len(power_readings)
    print(f"Average power usage over {duration} seconds: {average_power}W")

monitor_power_usage(60)  
```
